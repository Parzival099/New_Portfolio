# Análisis de los 25 retailers más grandes de Estados Unidos

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```
df = pd.read_csv("./largest_us_retailers.csv")
df.head()
```

```
df_sorted = df.sort_values('Sales', ascending=0)
df_sorted
```

## I. Preguntas del negocio

### 1. ¿Cuál es el promedio de ventas sin contar a la compañía dominante?

```
df_s = df["Sales"]
x = (df_s[1:])
Promedio=(x.sum()/len(x))
Promedio
```

```
x = (df["Sales"])
Promedio=(x.sum()/len(x))
Promedio
```

Segun el analisis el promedio sin contar la empresa dominante fue de 46,602, mientras que con ella fue de 71,063, Esto nos quiere decir que si un dato es demasiado grande el promedio se va a afectar y nuestro analisis estadistico no seria tan útil.

### 2. ¿Cuánto dinero en ventas generó la mayoría de las compañías?

```
df.Sales.hist()
```

La mayoria de compañias generaron menos de 100,000, mientras que aproximadamente 3 produjeron 100,000, y tambien hubo algunas empresas que facturaron aproximadamente 600,000.

### 3. ¿Cuántas tiendas tiene la mayoría de las compañías?

```
df.Stores.hist()
```

La mayoria de empresas se encuentran en menos de 4,00, y tambien hay algunas excepciones de mas de 4,000, algunas mas de 8,000, mientras hay algunas en 14,000 aproximadamente.

### 4. ¿La cantidad de tiendas está relacionada con la cantidad de ventas? Es decir, ¿si una empresa tiene más tiendas tendrá más dinero de ventas?

```
sns.scatterplot(data=df, x = 'Stores', y = 'Sales', hue = 'Company')
```

No, no esta directamente relacionado, porque la empresa con mas ventas, no posee pocas tiendas, ni la mayor cantidad de tiendas, y aun las empresas con la mayor cantidad de tiendas no llegan a las ventas de muchas empresas que tienen pocas ventas.

### 5. ¿Cuál es el rango que existe entre las ventas?

```
print("Este es el numero maximo en ventas")
print(df["Sales"].max())
print("Este es el numero minimo en ventas")
print(df["Sales"].min())
```

### 6. ¿Cuáles son las 5 empresas que más tiendas físicas tienen? ¿Cuáles de ellas están dentro de las 5 empresas que más ventas tuvieron?

```
x = df_sorted['Company'][0:5] #Aplicamos slicing como en una lista de Python.
y = df_sorted['Stores'][0:5]

plt.bar(x, y)
plt.title('Top 5 retailers')
plt.xlabel('Company')
plt.ylabel('Stores')
plt.xticks(rotation='vertical') #Método que se usa para rotar el texto de los puntos en X para que no se amontonen.
plt.show()
```

```
x = df_sorted['Company'][0:5] #Aplicamos slicing como en una lista de Python.
y = df_sorted['Sales'][0:5]

plt.bar(x, y)
plt.title('Top 5 retailers')
plt.xlabel('Company')
plt.ylabel('Sales')
plt.xticks(rotation='vertical') #Método que se usa para rotar el texto de los puntos en X para que no se amontonen.
plt.show()
```

De este analisis podemos concluir que Walgreen tiene la mayor cantidad de tiendas de los 5 primeros lugares seguido por Walmart y kroger, pero a su vez podemos concluir que dentro de las 5 estudiadas Walmart es la tienda con mayor cantidad de ventas entre las 5 analizadas, ademas fue la segunda en cantidad de tiendas, se podria decir que segun lo analizado Walmart es la mejor tienda tanto en ventas como en tiendas.
